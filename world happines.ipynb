```python
X=df[['Economy (GDP per Capita)', 'Family','Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']]
Y=df['Happiness Score']
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=20)
```

```python
y_train
```

```python
# step 1:
from sklearn.linear_model import LinearRegression
# step2:objext creation for algorith/model
model = LinearRegression()
model
```

```python
# step 3: training algprithm/model
model.fit(x_train,y_train)
```

```python
# step4 : model test
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
print('the model r2 score =',model.score(x_test,y_test)*100)
```
